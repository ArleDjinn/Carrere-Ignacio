```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import scipy 
import seaborn as sns
print('matplotlib.__version', mpl.__version__)
print('numpy.__version', np.__version__)
print('pandas.__version', pd.__version__)
print('scipy.__version', scipy.__version__)
print('seaborn.__version', sns.__version__)
```

```python
np.random.seed(3)
X = 3*np.random.rand(1000,1) # Crear 100 muestras de variable aleatoria de para la variable dependiente
                            # acuerdo a 3*Uniforme(0,1)
y = 5 + 2*X + np.random.randn(1000,1) # Crear 100 muestras para la variable dependiente
                                    # de acuerdo a 5 + 2X + uniforme(0,1)
                                    
                                    
                                    
X_ = np.c_[np.ones((1000,1)), X] #concatenar 1 a variables independientes
theta_opt = np.linalg.inv(X_.T.dot(X_)).dot(X_.T).dot(y) #calcular thetas de forma cerrada
theta_opt
```

```python
plt.plot(X,y, "b.")
plt.ylabel('y')
plt.xlabel('X')
plt.axis([0, 3, 0, 15])
plt.legend(["Regresión Lineal"])
plt.show()
```

```python
X_nuevo = np.array([[0], [3]])
print(X_nuevo)
X_nuevo_ = np.c_[np.ones((2,1)), X_nuevo]
print(X_nuevo_)
y_pred = X_nuevo_.dot(theta_opt)
print('pred', y_pred)
```

```python
plt.plot(X_nuevo, y_pred, 'r--')
plt.plot(X,y, "b.")
plt.ylabel('y')
plt.xlabel('X')
plt.axis([0, 3, 0, 15])
plt.legend(["Regresión Lineal"])
plt.show()
```
